Exploratory Data Analysis -- Day 4

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dataset/intern_data_cleaned.csv")
df.head()
```

```python
plt.figure()
plt.scatter(df["Tasks_Assigned"], df["Tasks_Completed"])
plt.xlabel("Tasks Assigned")
plt.ylabel("Tasks Completed")
plt.title("Tasks Assigned vs Completed")
plt.show()
```

```python
df.groupby("Department")["Performance_Score"].mean().plot(kind="bar")
plt.title("Average Performance by Department")
plt.show()
```

```python
plt.figure()
plt.hist(df["Attendance"])
plt.title("Attendance Distribution")
plt.show()
```

```python
plt.figure()
plt.hist(df["Completion_Rate"])
plt.title("Completion Rate Distribution")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
